# Linear Regression Demonstration

## **Linear Regression**

Linear reegression is the process of estimating the boundries in which data can lie on and make predictions on where the data points to. For this to occur we place some weights to measure how accurate we are in the hypothesis. This hypothesis is based on the function $h(x) = F(W,X)= \sum_{i=0}^n W_iX_i$ where the w are weights that are placed onto the x values to find the discriminant based classification.

This is setting up the data to be worked on. This is with the first number being one from how we need to times the add the first weight only and then multiply it onwards.

```python
data1 =[((1, -2), 1), ((1, -1), 3), ((1, 0), 5), ((1, 1), 7), ((1, 2), 9), ((1, 3), 11), ((1, 4), 13), ((1, 5), 15), ((1, 6), 17), ((1, 7), 19), ((1, 8), 21), ((1, 9), 23), ((1, 10), 25)]
data2 = [((1, -2), 6.39), ((1, -1), 16.51), ((1, 0), -3.11), ((1, 1), 10.79), ((1, 2), 11.62), ((1, 3), 23.24), ((1, 4), 18.27), ((1, 5), 27.58), ((1, 6), 22.21), ((1, 7), 5.12), ((1, 8), 8.86), ((1, 9), 10.69), ((1, 10), 14.82)]
```

This is setting up functions to be used for latter in the program. And will illustrate the function $h(x) = F(W,X)= \sum_{i=0}^n W_iX_i$.

```python
F = lambda W,X: sum([w*x for w,x in zip(W,X)])
```

This is setting up the x values that will be worked on and the expected value that function and weights would be predicting.   

```python
model1 = [0,0]
datax1= [d for d,_ in data1]
dataexpected1 = [p for _,p in data1]
```

This is the process of gathering the $F(W,X)$ and placing it in a list to better help with the programming

```python
ypredicted = [F(model1,x) for x in datax1]
```

## **Error within the linear regression**

This is to find the error within the linear algorithm function and to be used to see how much the weights work. This is found by the function $\sum_{i=0}^n (r-y)^2$ where $r$ is the expected value of the function and the $y$ is from the $h(x)$

```python
errorfunct = lambda dataexpe,ypredi: sum([(actual - predicted)**2 for actual,predicted in zip(dataexpe,ypredi)])
```

This is the error produced from the current weights being $W=[0,0]$

```python
error = errorfunct(dataexpected1,ypredicted)
print("error with ", model1, "\nError: ", error)
```

```python
dataX1 = [y for x, y in datax1]
plt.plot(dataX1, ypredicted, "g")
plt.plot(dataX1, dataexpected1, "g+")
plt.show()
```

## **Gradient descent**

Gradient descent is the process of updating the weights to better classify and predict the data. This is first done by finding the delta from the function of $d/dx_j = \frac{1}{n} \sum_{i=0}^n -2*(r-y)*x_j$. This would then be used to update the weight using the alpha to update the weight from the function $w_i = w_i - \alpha (d/dx_i)$ Then we continuously update it through various iterations to finely get adjust the weights to the right set. The alpha for this was chosen for being a good median of tryin to fit the weight into the least error.


Inital setup

```python
alpha = 0.01
iteration = 0
```

This is the first iteration 

```python
deltax0 = 1/len(dataexpected1) * sum([-2*(actual - predicted) * x[0] for actual,predicted,x in zip(dataexpected1,ypredicted,datax1)])
deltax1 = 1/len(dataexpected1) * sum([-2*(actual - predicted) * x[1] for actual,predicted,x in zip(dataexpected1,ypredicted,datax1)])
model1 = [model1[0] - (alpha * deltax0), model1[1] - (alpha*deltax1)]
iteration += 1
ypredicted = [F(model1,x) for x in datax1]
error = errorfunct(dataexpected1,ypredicted)
print("error with ", model1, "\nError: ", error," given the alpha of: ",alpha, "\non iteration: ",iteration)
```

```python
plt.plot(dataX1, ypredicted, "g")
plt.plot(dataX1, dataexpected1, "g+")
plt.show()
```

From this we can automate it to better suit the weights to accruatly predict what they will need to be for the right set adn lowest error. The number of iterations done here will be 100 for a good number and not to underfit the descent.

```python
for _ in range(100):
  ypredicted = [F(model1, x) for x in datax1]
  error = errorfunct(dataexpected1,ypredicted)
  if iteration%1==0:
    print("Iteration: ", iteration, "Error=", error)
    plt.plot(dataX1, ypredicted, "g")
    plt.plot(dataX1, dataexpected1, "g+")
    plt.show()
  deltax0 = 1/len(dataexpected1) * sum([-2*(actual - predicted) * x[0] for actual,predicted,x in zip(dataexpected1,ypredicted,datax1)])
  deltax1 = 1/len(dataexpected1) * sum([-2*(actual - predicted) * x[1] for actual,predicted,x in zip(dataexpected1,ypredicted,datax1)])
  model1 = [model1[0]-alpha*deltax0, model1[1]-alpha*deltax1]
  iteration += 1
```
